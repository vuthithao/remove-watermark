```python
import cv2
import numpy as np
import os

# Output = '/home/topica/Desktop'
img = cv2.imread("/home/topica/Downloads/67644505_639817889862371_7193897395610976256_n.jpg")
def watermask(img):
    gray  = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    dis = np.amax(gray) - np.amin(gray)
    color = img.mean(axis=0).mean(axis=0)
    alpha = 2.0
    beta = -160
    if sum(color/3) < 250 and dis > 190:
        new = alpha * img + beta
        new = np.clip(new, 0, 255).astype(np.uint8)
    else:
        new = img
    return new
new = watermask(img)
cv2.imwrite('r.jpg', new)
```

```python
Path = "/home/topica/Downloads/Media/Media/Problems"
folderlist = os.listdir(Path)
for i in range(len(folderlist)):
    try:
        imglist = os.listdir(os.path.join(Path, folderlist[i]))
        for j in range(len(imglist)):
                img = cv2.imread(os.path.join(Path, folderlist[i], imglist[j]))
                print(os.path.join(Path, folderlist[i], imglist[j]))
                new = watermask(img)
                cv2.imwrite(os.path.join(Path, folderlist[i], 'remove' + imglist[j]), new)
    except:
        continue
```
